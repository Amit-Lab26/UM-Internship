```python
# Create a new column 'Winner' based on the match result
matches_df['Winner'] = matches_df.apply(lambda row: row['Home Team Name'] if row['Home Team Goals'] > row['Away Team Goals'] else row['Away Team Name'], axis=1)

# Count the number of wins for each team
team_wins = matches_df['Winner'].value_counts()

# Select the top 5 champions with the highest number of wins
top_champions = team_wins.head(5)

# Visualize the top champions with the highest number of wins
sns.countplot(y=matches_df['Winner'], order=top_champions.index, palette="viridis")
plt.title('Top 5 Champions with Highest Number of Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.show()
```

```python
# Assuming cups_df is your DataFrame
team_wins = cups_df['Winner'].value_counts()
team_runners_up = cups_df['Runners-Up'].value_counts()

# Calculate the total and final counts
team_total = team_wins.add(team_runners_up, fill_value=0)
team_finals = team_wins.add(team_runners_up, fill_value=0)

# Create a DataFrame for visualization
plot_df = pd.DataFrame({'Winner': team_wins, 'Total': team_total, 'Final': team_finals}).reset_index()

# Melt the DataFrame for Seaborn bar plot
plot_df = plot_df.melt(id_vars='index', var_name='Category', value_name='Count')

# Plot the counts
plt.figure(figsize=(15, 8))
sns.barplot(x='index', y='Count', hue='Category', data=plot_df)
plt.title('Number of Times Each Team Has Been "Winner", "Total", and "Final"')
plt.xlabel('Team')
plt.ylabel('Count')
plt.legend(title='Category')
plt.show()
```

```python
# Number of Goals Scored
```

```python
# Combine home and away goals for each team
team_goals = pd.concat([matches_df.groupby('Home Team Name')['Home Team Goals'].sum(),
                       matches_df.groupby('Away Team Name')['Away Team Goals'].sum()], axis=1).fillna(0)

# Calculate the total goals for each team
team_goals['Total Goals'] = team_goals['Home Team Goals'] + team_goals['Away Team Goals']

# Find the team with the maximum total goals
max_goals_team = team_goals['Total Goals'].idxmax()
max_goals_count = team_goals['Total Goals'].max()

# Plot the total goals for each team
plt.figure(figsize=(15, 8))
team_goals['Total Goals'].sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Total Goals Scored by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.show()
```

```python
# Combine home and away goals for each player using MatchID
players_goals = players_df.groupby(['Player Name', 'MatchID'])[['Player Name']].count().rename(columns={'Player Name': 'Goals'}).reset_index()

# Calculate the total goals for each player
total_goals_by_player = players_goals.groupby('Player Name')['Goals'].sum()

# Sort players based on the number of goals
top_players = total_goals_by_player.sort_values(ascending=False).head(10)

# Plot the top 10 players with the most goals
plt.figure(figsize=(15, 8))
top_players.plot(kind='bar', color='salmon')
plt.title('Top 10 Players with the Most Goals')
plt.xlabel('Player')
plt.ylabel('Total Goals')
plt.show()
```

```python
# Combine home and away goals for each team
team_goals = pd.concat([matches_df.groupby('Home Team Name')['Home Team Goals'].sum(),
                       matches_df.groupby('Away Team Name')['Away Team Goals'].sum()], axis=1).fillna(0)

# Calculate the total goals for each team
team_goals['Total Goals'] = team_goals['Home Team Goals'] + team_goals['Away Team Goals']

# Sort teams based on the total goals
team_goals_sorted = team_goals['Total Goals'].sort_values(ascending=False)

# Plot the total goals scored by each team
plt.figure(figsize=(15, 8))
team_goals_sorted.plot(kind='bar', color='skyblue')
plt.title('Total Goals Scored by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.show()
```

```python
#calculate and visualize the average goals per match for each team:
```

```python
# Calculate the total matches played by each team
matches_played = pd.concat([matches_df['Home Team Name'], matches_df['Away Team Name']]).value_counts()

# Calculate the average goals per match for each team
average_goals_per_match = team_goals['Total Goals'] / matches_played

# Sort teams based on average goals per match
average_goals_sorted = average_goals_per_match.sort_values(ascending=False)

# Plot the average goals per match for each team
plt.figure(figsize=(15, 8))
average_goals_sorted.plot(kind='bar', color='lightgreen')
plt.title('Average Goals per Match for Each Team')
plt.xlabel('Team')
plt.ylabel('Average Goals per Match')
plt.show()
```
